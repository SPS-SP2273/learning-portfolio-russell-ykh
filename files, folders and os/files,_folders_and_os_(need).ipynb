### 2.4 Checking for existence

```python
import os

for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    try:
        os.mkdir(path)
        print(f'Creating {path}')
    except FileExistsError:
        print(f'{path} already exists; skipping creation.')

for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    if os.path.exists(path):
        print(f'{path} already exists; skipping creation.')
    else:
        os.mkdir(path)
        print(f'Creating {path}')
```

### 2.5 Copying files

```python
import os
import shutil

for person in ['John', 'Paul', 'Ringo']:
    path_to_destination = os.path.join('people', person)
    shutil.copy('sp2273_logo.png', path_to_destination)
    print(f'Copied file to {path_to_destination}')
    
for person in ['John', 'Paul', 'Ringo']:
    # Create folder 'imgs'
    path_to_imgs = os.path.join('people', person, 'imgs')
    if not os.path.exists(path_to_imgs):
        os.mkdir(path_to_imgs)

    # Move logo file
    current_path_of_logo = os.path.join('people', person, 'sp2273_logo.png')
    new_path_of_logo = os.path.join('people', person, 'imgs', 'sp2273_logo.png')

    shutil.move(current_path_of_logo, new_path_of_logo)
    print(f'Moved logo to {new_path_of_logo}')
```

## Exercise 7 :  The OS is my oyster

1. Use os.getcwd() to figure out where you are?
1. Use os.mkdir() to create a new directory (folder) named python_today and use os.chdir() to change into this directory.
 * Verify that you are where you want to be.
1. Use os.mkdir() (along with for) to create 50 new directories named in the format new_directory_01, new_directory_02, …
1. Use os.rmdir() to remove all the odd numbered new directories.
 * Use os.listdir() to check your handiwork.
1. Use os.rename() to rename every 5th (i.e. 5, 10, 15, …) new_directory to haha_05, haha_10, haha_15, …
 * Use os.path.exist() to account for missing files.
 * Use os.listdir() to check your handiwork.
1. Copy the module logo (in the main folder of the Learning Portfolio) to all the haha folders.
 * Use glob to get the details of haha folders.
 * Use shutil.copy() to copy the files
 * Remember to use either os.path.sep or os.path.join()
 * Use os.listdir() to check your handiwork.

```python
import os
import shutil
import glob

def default_path():
    return os.path.join("/Users","Russell","Documents","GitHub","learning-portfolio-russell-ykh","files, folders and os")

def m(path):
    if not(os.path.exists(path)):
        os.mkdir(path)
        
def d(num):
    return "new_directory_" + str.zfill(str(num),2)

#1
os.chdir(default_path())
print(os.getcwd())
if os.path.exists("python_today"):
    shutil.rmtree("python_today")

#2
new_path = "python_today"
m(new_path)
os.chdir(new_path)
print(os.getcwd())

#3
for i in range(1, 51):
    m(d(i))  
print(glob.glob("*",recursive=True))

#4 + 5
for i in range(1, 51):
    if i % 2 == 1:
        os.rmdir(d(i))
    elif i % 5 == 0:
        os.rename(d(i),"haha_"+str.zfill(str(i),2))

print(os.listdir())

#6
os.chdir(default_path())
img = "sp2273_logo.png"
hahas = glob.glob(os.path.join("python_today","haha*"))

for haha in hahas:
    shutil.copy(img,haha)

print(glob.glob(os.path.join("python_today","haha*/*"),recursive=True))
```

## Exercise 8 :  Tidying a collaboration

```python
import os
import glob

months = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

#old name is in format month-date_city.smth
def rename_my_file(old):
    month = old[0:old.index("-")]
    day = old[old.index("-")+1:old.index("_")]
    city = old[old.index("_")+1:old.index(".")]
    ext = old[old.index("."):]
    new = months[month] + "-" + day + "-" + city + ext
    return new

def rename_test():
    old_files = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

    new_files = []

    for f in old_files:
        new_files.append(rename_my_file(f))
        
    for n in new_files:
        print(n)
        print(extract_city(n))
    
def rename_all_txt():
    txts = glob.glob(os.path.join("os-collaboration-exercise-data", "*.txt"))
    
    for txt in txts:
        f = os.path.split(txt)[1]
        
        #if the file is in the old format - months as words - then rename
        if(not(f[0] in [str(i) for i in range(10)])):
            new_name = rename_my_file(f)
            os.rename(txt, os.path.join("os-collaboration-exercise-data", new_name))
    
    print(txts)
    
def extract_city(data):
    txt = os.path.split(data)[1]
    city = txt[txt.rindex("-")+1:txt.index(".")]
    city_cap = city[0].upper() + city[1:]
    return city_cap

def move_all_txt():
    txts = glob.glob(os.path.join("os-collaboration-exercise-data", "*.txt"))
    
    for city in cities:
        try:
            os.mkdir(city)
        except FileExistsError:
            print("This folder already exists... Skipping...")
    
    for txt in txts:
        city = extract_city(txt)
        #shutil.copy(txt, city)
        shutil.move(txt, city)

rename_all_txt()
move_all_txt()
```
