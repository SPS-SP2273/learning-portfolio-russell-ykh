### 1.7 We need another axis!

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0,6,30)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dotted', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

## Exercise 1 :  Fill in the blanks

```python
from matplotlib import pyplot as plt
import numpy as np

# Some data for plotting
x = np.arange(0,6)
y = x.copy() * 4
x_err = 0.25
y_err = x.copy() * 0.5

plt.plot(x, y)
plt.style.use("default")
plt.errorbar(x, y, xerr=x_err, yerr=y_err,
             color="darkred", linestyle="dashed",
             marker="o", markerfacecolor="yellow",
             ecolor="black",capsize=5,
             label="Y values with errors")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc="lower right")
plt.show(block=False)
```

## Exercise 2 :  A simple mathematical plot

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0, 50, 200)
ex = np.e ** (-1 * x.copy() / 10)
sinx = np.sin(x.copy())
y = ex * sinx
plt.plot(x, y)
plt.grid(alpha=0.25)
plt.xlabel("X")
plt.ylabel("Y",rotation=0)
plt.title("$y = e^{-x/10} sin(x)$")
plt.ylim(-1, 1)
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0, 50, 200)
ex = np.e ** (-1 * x.copy() / 10)
sinx = np.sin(x.copy())
y = ex * sinx
plt.plot(x, y)
plt.grid(alpha=0.25)
plt.xlabel("x", fontsize=15, color="darkblue")
plt.ylabel("y", rotation=0, fontsize=15, color="darkblue")
plt.title("$y = e^{-x/10} sin(x)$", fontsize=20)
plt.ylim(-1, 1)
plt.show()
```
